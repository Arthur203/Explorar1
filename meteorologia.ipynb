```python
import pandas as pd
import seaborn as srn 
import statistics  as sts
import matplotlib.pyplot as plt
```

```python
base = pd.read_csv('trabalho meteoro.csv', sep = ";")
base.head()
```

```python
# Supondo que você tenha lido os dados em um DataFrame chamado 'df'
# Substitua as vírgulas por pontos nas colunas numéricas
base['ARM'] = base['ARM'].replace(',', '.', regex=True)
base['ETR'] = base['ETR'].replace(',', '.', regex=True)
base['Defict'] = base['Defict'].replace(',', '.', regex=True)
base['Excedente'] = base['Excedente'].replace(',', '.', regex=True)
base['Eto'] = base['Eto'].replace(',', '.', regex=True)
base['PRECIPITACAO'] = base['PRECIPITACAO'].replace(',', '.', regex=True)
base['TEMPERATURA'] = base['TEMPERATURA'].replace(',', '.', regex=True)



```

```python
base['PRECIPITACAO'] = base['PRECIPITACAO'].astype(float)
```

```python
agrupado = base.groupby(["DATA"])["PRECIPITACAO"].sum()
agrupado 
```

```python
colors = ['royalblue', 'mediumseagreen', 'darkorange', 'indianred', 'mediumorchid', 'gold', 'cornflowerblue', 'tomato', 'mediumaquamarine', 'plum', 'lightcoral', 'cadetblue']
agrupado.plot.bar(color=colors)

plt.title('PRECIPITAÇÃO NO ANO')
plt.show()
```

```python
base['TEMPERATURA'] = base['TEMPERATURA'].astype(float)
```

```python
agrupado = base.groupby(["DATA"])["TEMPERATURA"].sum()
agrupado 
```

```python
colors = ['dodgerblue', 'limegreen', 'orangered', 'darkviolet', 'gold', 'teal', 'salmon', 'mediumspringgreen', 'mediumslateblue', 'darkkhaki', 'mediumturquoise', 'palevioletred']
agrupado.plot.bar(color=colors)

plt.title('TEMPERATURA NO ANO')
plt.show()
```
